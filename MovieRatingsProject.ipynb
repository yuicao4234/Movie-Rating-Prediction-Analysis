Here we do PCA to start the dimension reduction method. 

```python
from scipy import stats 
from sklearn.decomposition import PCA

# zscore data

sensation_zscore = stats.zscore(sensation)

# direct PCA
pca = PCA().fit(sensation_zscore)

eigVals = pca.explained_variance_

loadings = pca.components_

varExplained = eigVals/sum(eigVals)*100 # 20 questions, so eigsum should be 20

# Now let's display this for each factor:
for ii in range(len(varExplained)):
    print(varExplained[ii].round(3))
```

```python
# Scree Plot
numQuestions = len(sensation.columns)
x = np.linspace(1,numQuestions,numQuestions) 
plt.bar(x, eigVals, color='gray')
plt.plot([0,numQuestions],[1,1],color='orange') # Orange Kaiser criterion line for the fox
plt.xlabel('Principal component')
plt.ylabel('Eigenvalue')
plt.show()
```

If we use Kaiser Threshold, we will have 5 PCs. 
If we used elbow threshold, we'll only use 1. 

Try using threshold = 95% of variance. 


```python
# Testing out how many factors we'll be using for each threshold. 
# set threshold to 90%. 

# Kaiser 
kaiserThreshold = 1 
print('Number of factors selected by Kaiser criterion:', np.count_nonzero(eigVals>kaiserThreshold))

# elbow: one factor
print('Number of factors selected by elbow criterion: 1')


threshold = 90
eigSum = np.cumsum(varExplained)
print('Number of factors to account for at least 90% variance:', np.count_nonzero(eigSum < threshold) + 1)
```

Now let's look at the Loading Score. Based on Kaiser Criterion, let's look at the first 7 PCs. 

```python
fig, axes = plt.subplots(2, 4, figsize=(24, 8), sharey=True)
axes = axes.flatten() 
pcs = range(7)

for i, pc in enumerate(pcs):
    ax = axes[i]
    ax.bar(x, loadings[pc, :] *-1)
    ax.set_title(f'PC{pc + 1}')
    ax.set_xlabel('Question')
    ax.set_xticks(x)
    ax.tick_params(axis='x')
    
axes[-1].axis('off')

axes[0].set_ylabel('Loading')

plt.tight_layout()
plt.show()

print(labels)
```

Looking at each loading plots: 

PC1: Most questions have about the same magnitude hovering at around 0.35. PC1 splits the sensation seeking questions into 2 major directions: either thrill-seeking or cautious lifestyle. For Q 15-17, the direction goes upward, and these are more distinct questions about lifestyles compared to the others (which are similar in terms of thrill-seeking). Let's call this PC1 "Lifestyle". 

PC2: Thrilling activity / "Impulsivity".

PC4: Opposite direction between Q6 and Q17, showing characteristics on ordering, planning. "Order and Control".

### Hypothesis Test

H0: Sensation_PC[i] is not related to Movie Experience[j].

H1: Higher Sensation_PC[i] is associated with higher Movie Experience[j]. 

We reject the null hypothesis at the  = 0.005 significance level.

```python
from scipy.stats import pearsonr

results = []

for ss in sensation_pca.columns:
    for me in movieexp_pca.columns:
        r, p = pearsonr(sensation_pca[ss], movieexp_pca[me])
        results.append({
            "Sensation_PC": ss,
            "Movie_PC": me,
            "r": r,
            "p": p
        })

results_df = pd.DataFrame(results)
results_df.sort_values("p")
```

Based on the pearson correlation coefficient test, we can conclude: 

- for Life Stability, p-value is statistically significant at alpha = 0.005, and we reject the H0 that there is no relationship between Life Stability and Movie Imemrsiveness. Life stability is positively associated with movie immersiveness. 

- for Order & Control, Impulsivity, and Fear Tolerance, p-value is statistically significant at alpha = 0.005, and we reject the H0 that there is no relationship between these variables and Movie Immersiveness. These variables are negatively associated with movie immersiveness. 


## 2. Is there evidence of personality types based on the data of these research participants? If so, characterize these types both quantitatively and narratively.

K-Means + PCA

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples


# Store PCA vector as X
X = personality_pca.values

# start with numclusters = 10
numClusters = 10
sMean = np.full((numClusters-1, 1), np.nan)  # indices 0..8 correspond to k=2..10

for k in range(2, numClusters + 1):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X)
    labels = kmeans.labels_

    # per-point silhouettes (for histogram)
    s = silhouette_samples(X, labels)

    # standard summary metric: mean silhouette score
    sMean[k-2] = s.mean()

    # Plot histogram of silhouettes for this k
    plt.subplot(3, 3, k-1)
    plt.hist(s, bins=20)
    plt.xlim(-0.2, 1)
    plt.ylim(0, 250)
    plt.xlabel("Silhouette score")
    plt.ylabel("Count")
    plt.title(f"k={k}, mean={s.mean():.3f}")
    plt.tight_layout()

plt.show()

# Optional: line plot to pick k
plt.plot(range(2, numClusters+1), sMean.flatten(), marker="o")
plt.xlabel("k")
plt.ylabel("Mean silhouette score")
plt.title("Silhouette vs k")
plt.show()
```

```python
#%% 2j) Now that we determined the optimal k, we can now ask kMeans to cluster the data for us, 
#assuming that k

# kMeans:
numClusters = 2
kMeans = KMeans(n_clusters=numClusters, random_state=42, n_init=10).fit(X)
cId = kMeans.labels_
cCoords = kMeans.cluster_centers_

for ii in range(numClusters):
    idx = np.where(cId == ii)[0]
    plt.plot(X[idx, 0], X[idx, 1], 'o', markersize=1)
    plt.plot(cCoords[ii, 0], cCoords[ii, 1], 'o', markersize=6, color='black')

plt.xlabel('PC1: Social Energy')
plt.ylabel('PC2: Emotional Stability')
plt.title('K-means Clusters in PCA Space')
plt.show()
```

Narratively: Based on PCA and K-Means clustering, we can see that there are 2 types of personality: Low social energy vs High social energy. 

Quantitatively: 
(1) Reduced dimension with PCA, indicating 6 principle components based on Kaiser criterion (factors account for >90% of eigensum). I named these principal components as: 
PC1. "Social Energy"
PC2. "Emotional Stability"
PC3. "Independence"
PC4. "Behavioral Discipline"
PC5. "Interpersonal & Altruistic"
PC6. "Artistic Interest"

(2) K-Means Clustering: I used the 6 principle components from the above and conducted K-Means Clustering. Using Silhouette Analysis, k=2 was selected as the optimal number of k. For these 2 clusters, each cluster center represents: i. relatively low social energy & neutral emotional stability, and ii. relatively high social energy and neutral emotional stability. 

## 3.Are movies that are more popular rated higher than movies that are less popular?

Let's define "popular" first. We define "Popularity" with response rate (responses that are not NaNs) higher than a threshold. 
We first start by looking at the overall response rate. This question will require us to go back to the original, "uncleaned" dataset. 

```python
# Response Rate 
# read the raw dataset
df_raw = pd.read_csv("movieReplicationSet - Copy.csv")

response_rate = df_raw.iloc[:,:400].notnull().sum().sum() / df_raw.size
print(response_rate)
```

```python
# dividing movies based on their popularity / number of responses
movies_raw = df_raw.iloc[:, :400] 

movie_medians = movies_raw.median(axis=0, skipna=True)

# response rate for each column
response_rate = movies_raw.notnull().sum() / movies_raw.shape[0]
median_response_rate = response_rate.median()

popular_idx = response_rate[response_rate >= median_response_rate].index
niche_idx = response_rate[response_rate < median_response_rate].index

popular_ratings = movies_raw[popular_idx].values.flatten()
niche_ratings   = movies_raw[niche_idx].values.flatten()

# remove null values
popular_ratings = popular_ratings[~np.isnan(popular_ratings)]
niche_ratings = niche_ratings[~np.isnan(niche_ratings)]
```

```python
from scipy.stats import mannwhitneyu

# Mann Whitney U Test
u_stat, p_value = mannwhitneyu(
    popular_ratings,
    niche_ratings,
    alternative="greater"  # popular rated higher than niche
)

print("U:", u_stat)
print("p:", p_value)
```

```python
alpha = 0.005

print("H0: The distribution of ratings for popular movies is not greater than that of niche movies.")
print("H1: The ratings for popular movies are higher than those of niche movies.")

if p_value < alpha:
    print("Reject H0.")
else:
    print("Fail to reject H0.")
```

p < 0.005, so we can reject the null hypothesis. Popular movies are rated higher than less popular movies. 

## 4. Is enjoyment of Shrek (2001) gendered, i.e. do male and female viewers rate it differently?

```python
# Hypothesis: Ratings for Shrek between Male and Female viewers are different. 
# here we will continue using the raw dataset. 

gender_col = df_raw.columns[-3]
female_col = df_raw.loc[df_raw[gender_col] == 1]
male_col = df_raw.loc[df_raw[gender_col] == 2] 

shrek_female = female_col["Shrek (2001)"].dropna()
shrek_male = male_col["Shrek (2001)"].dropna()

# check sample sizes 
print(shrek_female.size)
print(shrek_male.size)
```

```python
# Mann Whitney U Test
u_stat, p_value = mannwhitneyu(
    shrek_female,
    shrek_male,
    # nan_policy = 'Omit': omit nans
    alternative= 'two-sided'  # testing whether distributions are equal. 
    # The H0 will be: Two distributions are equal. 
)

print("U:", u_stat)
print("p:", p_value)
```

```python
alpha = 0.005

print("H0: There is no difference in the ratings for Shrek (2001) between female viewers and male viewers.")
print("H1: Ratings for Shrek (2001) differ between female and male viewers.")

if p_value < alpha:
    print("Reject H0.")
else:
    print("Fail to reject H0.")
```

We fail to reject null hypothesis that there is no difference in the ratings for "Shrek (2001)" between female viewers and male viewers.

## 7. There are ratings on movies from several franchises ([Star Wars, Harry Potter, The Matrix, Indiana Jones, Jurassic Park, Pirates of the Caribbean, Toy Story, Batman]) in this dataset. How many of these are of inconsistent quality, as experienced by viewers?


To answer this question, we use Kruskal-Wallis Test to determine if the median is consistent across different numberings for each franchise (difference in mean ranks).

```python
# Franchise movies

franchise = ["Star Wars", "Harry Potter", "The Matrix", "Indiana Jones", "Jurassic Park",
            "Pirates of the Caribbean", "Toy Story", "Batman"]
franchise = {name: [] for name in franchise}

# for i in df, if title includes franchise name, add col name under the franchise
for key in franchise:
    for col in df_raw.columns:
        if key in col:
            franchise[key].append(col)

franchise
```

```python
# do kruskal test for each franchise 
results = {}

for title, numbering in franchise.items(): 
    movie = [] 
    for number in numbering:
        vals = df_raw[number].dropna().values 
        movie.append(vals)

    Hval, pval = stats.kruskal(*movie)
    results[title] = {"H": Hval, "p": pval, "n_groups": len(movie)}
```

```python
for name, res in results.items():
    print(name, "-> H:", res["H"], ", p:", res["p"], ", groups:", res["n_groups"])
    if res["p"] < 0.05: 
        print("Reject H0: conclude that ratings are not consistent.")
    else:
        print("Fail to reject H0: conclude that ratings are consistent.")
    print()
```

```python
# Visualize median rating for each movie

n_franchises = len(franchise)
n_cols = 4
n_rows = (n_franchises + n_cols - 1) // n_cols   # ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 8), sharey=True)
axes = axes.flatten()   # make it easy to index

# 2. Loop through each franchise and plot median rating per movie
for i, (name, cols) in enumerate(franchise.items()):
    ax = axes[i]
    
    # take only those columns and compute medians
    medians = df_raw[cols].median(axis=0, skipna=True)

    ax.bar(range(len(medians)), medians.values)
    ax.set_title(name)
    ax.set_xticks(range(len(medians)))
    ax.set_xticklabels(medians.index, rotation=60, ha="right", fontsize=8)
    ax.set_ylabel("Median rating")

# 3. Hide any unused subplots (if franchises not multiple of 4)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Based on the test results, we can see that all franchises except for Star Wars have inconsistent ratings.
